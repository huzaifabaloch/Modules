# Sorting/Describing Data

```python
df.describe()
```

```python
# Sorting single column ASC and DESC
df.sort_values('Name', ascending=True)[:5]
df.sort_values('Name', ascending=False)[:5]
```

```python
# Sorting multiple columns 1 -> asc, 2 -> desc
df.sort_values(['Type 1', 'HP']).head()
df.sort_values(['Type 1', 'HP'], ascending=[1, 0]).head()
```

# Making changes to the data

```python
# Making a new column that is the sum of all columns, just to get the statistic of each pokemon.
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
df.head()
```

```python
# Verifying
45+49+49+65+65+45
```

```python
# reseting the dataframe
df = df.drop(columns=['Total']) 
```

```python
# Second way
# iloc [select_all_rows, number_of_columns].sum(horizontally)
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
```

```python
df.head()
```

```python
# reordering columns
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head()
```

# Saving our data (EXPORTING INTO DESIRED FORMAT)
